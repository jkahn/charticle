```python
%matplotlib inline

from matplotlib import pyplot as plt
from charticle.venn import Venn3
```

```python
v3 = Venn3(a_name="useful", b_name = "structured", c_name="delimited")
v3.abc = "discipline"
v3.ab = "craft"
v3.sizes.abc *= 3
v3.plot(); plt.show()
```

```python
v3.a, v3.b, v3.c = "insight", "code of\nbehavior", "procedure for\nvalidation"
v3.palette.a = 'orange' # leave green, blue
v3.palette.alpha = 0.7
v3.plot(); plt.show()
```

```python
import attr
v3.palette.alpha=-0.1
try:
    attr.validate(v3.palette)
    assert "validation should fail"
except ValueError:
    pass
    v3.palette = v3.Palette() # reset
```
